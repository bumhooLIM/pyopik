```python
import numpy as np
from pyopik import opik_probability
```

```python
# Example calculation following Table 1 in Dell'Oro & Paolicchi 1998
a, e, i = 1.59, 0.056, 0.466
a0, e0, i0 = 3.42, 0.578, 0.435

P_i, U = opik_probability(
    a=a, e=e, i=i,    # Projectile
    a0=a0, e0=e0, i0=i0, # Target
    N=10000, return_velocity=True
)
print(f"Probability: {P_i} km^-2 yr^-1")
print(f"Mean Impact Velocity: {U} km s^-1")
```

```python
# Swap target <-> projectile

a, e, i = 3.42, 0.578, 0.435
a0, e0, i0 = 1.59, 0.056, 0.466

P_i, U = opik_probability(
    a=a, e=e, i=i,    # Projectile
    a0=a0, e0=e0, i0=i0, # Target
    N=100000, return_velocity=True
)
print(f"Probability: {P_i} km^-2 yr^-1")
print(f"Mean Impact Velocity: {U} km s^-1")
```
